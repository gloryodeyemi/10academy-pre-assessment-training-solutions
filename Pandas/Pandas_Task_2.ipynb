#### Question 1. Import the **pandas_tasks.csv** file into the notebook and view the head, tail and a sample of 10 random rows.

```python
#importing the library
import pandas as pd

#importing the dataset
data = pd.read_csv('pandas_tasks.csv')

#viewing the head
print("    Head \n-----------")
data.head()

#viewing the tail
print("    Tail \n-----------")
data.tail()

#viewing a sample of 10 random rows
print(" Sample of 10 Random rows \n---------------------------- \n")
data.sample(10)
```

#### Question 2. Sort the dataset according to Q4, Q5, Q3 columns separately

```python
#sorting the dataset according to Q4
sort1 = data.sort_values(by=['Q4'])
sort1

#sorting the dataset according to Q5
sort2 = data.sort_values(by=['Q5'])
sort2

#sorting the dataset according to Q3
sort3 = data.sort_values(by=['Q3'])
sort3
```

#### Question 3. Select the first 10 columns, the last 10 columns, and the 12th-19th columns.

```python
#Selecting the first 10 columns
col1 = data.iloc[:, 0:10]
col1

#Selecting the last 10 columns
col2 = data.iloc[:, -10:]
col2

#Selecting the 12th-19th columns
col3 = data.iloc[:, 12:20]
col3
```
